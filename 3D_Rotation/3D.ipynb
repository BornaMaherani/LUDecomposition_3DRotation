```python
import matplotlib.pyplot as pyplot
import numpy as np
import math
from mpl_toolkits import mplot3d

# Before Plot
fig = pyplot.figure(figsize = (8, 8))
ax = pyplot.axes(projection = '3d')
# After Plot 
fig2 = pyplot.figure(figsize = (8, 8))
ax2 = pyplot.axes(projection = '3d')

# rotation axis
Axis = input("Enter the axis of rotation(x,y or z): ")
# angle of rotation
Angle = int(input("Enter the angle of rotation in degrees: "))
# number of dots in space 
n = int(input("How many 3D point will ypu input: "))

# calculating cos and sin of the rotation degree
cos = math.cos(math.radians(Angle))
sin = math.sin(math.radians(Angle))
# which rotation matrix?
if Axis == 'x':
    A = np.array([
        [1, 0, 0],
        [0, cos, -sin],
        [0, sin, cos]
    ])
elif Axis == 'y':
    A = np.array([
        [cos, 0, sin],
        [0, 1, 0],
        [-sin, 0, cos]
    ])
elif Axis == 'z':
    A = np.array([
        [cos, -sin, 0],
        [sin, cos, 0],
        [0, 0, 1]
    ])

# draw dots before and after rotation in 3d space
max_size = 0
for i in range (n):

    # Before
    point = input("Enter the coordinates of each point as x,y,z (seperated by commas): ").split(",")
    point = list(map(int, point))

    #label of a dot
    label = '(%d, %d, %d)' % (point[0], point[1], point[2])
    ax.scatter(point[0], point[1], point[2], c='Red')
    ax.text(point[0], point[1], point[2], label)

    # After
    point = np.dot(A, point)

    #label of a dot
    label = '(%d, %d, %d)' % (point[0], point[1], point[2])
    ax2.scatter(point[0], point[1], point[2], c='Blue')
    ax2.text(point[0], point[1], point[2], label)

    #for scaling the plot
    max_size = max(max_size, point[0], point[1], point[2])

ax.set_xlim(0, max_size)
ax.set_ylim(0, max_size)
ax.set_zlim(0, max_size)

ax2.set_xlim(0, max_size)
ax2.set_ylim(0, max_size)
ax2.set_zlim(0, max_size)


ax.set_title('3D plot: Before rotation')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax2.set_title('3D plot: After rotation')
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')
ax2.set_zlabel('Z Axis')
pyplot.show()
```
